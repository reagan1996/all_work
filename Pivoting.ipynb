```python
# pd.options.display.float_format = '{:,.2f}'.format
pd.options.display.float_format = '{0:.0%}'.format
```

```python
xlsfile = pd.ExcelFile("Documents/Jessample.xlsx")
```

```python
dframe = xlsfile.parse("Jessample")
```

```python
dframe
```

```python
dframe_piv = dframe.pivot("date", "Salesperson", "Total Amount")
dframe_piv
```

```python
dframe_piv_pct_change = dframe_piv.pct_change()
dframe_piv_pct_change
```

```python
dframe_piv_pct_change = dframe_piv_pct_change.replace(np.nan, '-', regex=True)
dframe_piv_pct_change
```
