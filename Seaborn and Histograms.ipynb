# Regression Plots

```python
dataset=pd.read_csv("Documents/Weather_data.csv")
```

```python
sns.lmplot("Temperature", "Wind speed", dataset)
```

```python
sns.lmplot("Temperature", "Wind speed", dataset,
          scatter_kws={"marker":"o", "color":"indianred"},
          line_kws={"linewidth":4, "color":"blue"})
```

```python
sns.lmplot("Temperature", "Wind speed", dataset, order=2,
          scatter_kws={"marker":"o", "color":"indianred"},
          line_kws={"linewidth":4, "color":"blue"})
```

```python
sns.lmplot("Temperature", "Wind speed", dataset, fit_reg=False)
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.tail()
```

```python
tips["tips_pct"]= 100*(tips["tip"]/tips["total_bill"])
```

```python
tips.head()
```

```python
sns.lmplot("size", "tips_pct", tips, x_estimator=np.mean)
```

```python
sns.lmplot("size", "tips_pct", tips, hue= "sex", markers=["x", "o"])
```

```python
sns.lmplot("size", "tips_pct", tips, hue= "day")
```

```python
sns.regplot("total_bill", "tips_pct", tips)
```

# Heat maps and Clustered matrices

```python
flight_dframe = sns.load_dataset("flights")
flight_dframe.head()
```

```python
flight_dframe = flight_dframe.pivot("month", "year", "passengers")
flight_dframe
```

```python
# creating heatmaps
```

```python
sns.heatmap(flight_dframe)
```

```python
sns.heatmap(flight_dframe, annot=True, fmt="d")
```

```python
sns.heatmap(flight_dframe, center = flight_dframe.loc["January", 1955])
```

```python
# clustered matrices
```

```python
sns.clustermap(flight_dframe)
```

```python
sns.clustermap(flight_dframe, col_cluster=False)
```

```python
sns.clustermap(flight_dframe, standard_scale=1)
```

```python
sns.clustermap(flight_dframe, standard_scale=0)
```
